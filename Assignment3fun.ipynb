1 What is the difference between a function and a method in Python?
- Function:

A function is a block of code that performs a specific task.
It is defined using the def keyword.
It can be called independently.
Example:

def greet(name):
  print(f"Hello, {name}!")

greet("Alice")
- Method:

A method is a function that is associated with an object.
It is defined within a class.
It is called on an object of that class.
Example:

class Dog:
  def bark(self):
    print("Woof!")

my_dog = Dog()  # Creating an object of the class
my_dog.bark()   # Calling the method on the object

 2 Explain the concept of function arguments and parameters in Python.

- Parameters

Parameters are the names listed in the function definition.
They act as placeholders for the values that will be passed to the function when it's called.
Think of them as variables that are assigned values within the function's scope.
- Arguments

Arguments are the actual values passed to a function when it's called.
They provide the concrete data that the function will work with. They correspond to the parameters defined in the function's signature.

3 What are the different ways to define and call a function in Python?

- In Python, functions are a central part of the language, and there are several ways to define and call them. Below are the primary types:

---

### 1. **Regular Function**
#### Definition:
A standard function is defined using the `def` keyword, followed by a name, parentheses, and a colon. Optionally, it may include parameters.

```python
def greet(name):
    return f"Hello, {name}!"
```

#### Call:
```python
print(greet("Alice"))  # Output: Hello, Alice!
```

---

### 2. **Function with Default Parameters**
#### Definition:
Functions can have parameters with default values, which are used if no arguments are provided during the call.

```python
def greet(name="Guest"):
    return f"Hello, {name}!"
```

#### Call:
```python
print(greet())           # Output: Hello, Guest!
print(greet("Alice"))    # Output: Hello, Alice!
```

---

### 3. **Lambda Function (Anonymous Function)**
#### Definition:
A lambda function is a compact way to define a function without a name, typically for simple, one-liner operations.

```python
add = lambda x, y: x + y
```

#### Call:
```python
print(add(2, 3))  # Output: 5
```

---

### 4. **Nested Function**
#### Definition:
A function defined within another function.

```python
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function
```

#### Call:
```python
add_five = outer_function(5)
print(add_five(3))  # Output: 8
```

---

### 5. **Recursive Function**
#### Definition:
A function that calls itself to solve a smaller instance of the same problem.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

#### Call:
```python
print(factorial(5))  # Output: 120
```

---

### 6. **Method (Function in a Class)**
#### Definition:
Functions defined inside a class are methods. They often take `self` as the first parameter.

```python
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"
```

#### Call:
```python
greeter = Greeter()
print(greeter.greet("Alice"))  # Output: Hello, Alice!
```

---

### 7. **Built-in Functions**
Python provides many built-in functions like `len()`, `print()`, etc. These are called directly.

#### Call:
```python
print(len("Hello"))  # Output: 5
```

---

### 8. **Decorator Functions**
#### Definition:
A decorator is a higher-order function that modifies or extends the behavior of another function.

```python
def decorator(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper

@decorator
def say_hello():
    print("Hello!")
```

#### Call:
```python
say_hello()
# Output:
# Before the function call
# Hello!
# After the function call
```

---

### 9. **Callable Objects**
#### Definition:
Instances of a class that implements the `__call__` method can be called like functions.

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    
    def __call__(self, x):
        return x * self.factor
```

#### Call:
```python
double = Multiplier(2)
print(double(5))  # Output: 10
```

---

### 10. **Asynchronous Functions**
#### Definition:
Defined using `async def`, these are used for asynchronous programming and typically involve `await`.

```python
import asyncio

async def greet_async(name):
    await asyncio.sleep(1)
    return f"Hello, {name}!"
```

#### Call:
```python
asyncio.run(greet_async("Alice"))  # Output: Hello, Alice!
```

---

By combining these techniques, you can create a variety of functions for different use cases in Python.

6  Explain the concept of generators in Python and how they are defined.

- Generators are a special type of iterable in Python, used to produce a sequence of values lazily, one at a time, instead of generating all values at once and storing them in memory. They are especially useful when working with large datasets or streams of data.

Key Characteristics of Generators
Lazy Evaluation:

Generators produce values only when requested, conserving memory.
They generate values on-the-fly rather than precomputing and storing them.
Stateful Iteration:

Generators maintain their state between iterations, allowing them to "remember" where they left off.
No Memory Overhead:

Unlike lists or other data structures, generators don't store all values in memory at once.
How Generators Are Defined
Generators can be created in two main ways:

- 1. Using yield in a Function
A generator function is defined like a regular function but uses the yield keyword instead of return. Each call to yield produces a value and pauses the function, saving its state for the next iteration.
- 2. Using Generator Expressions
Generator expressions are a concise way to create generators, similar to list comprehensions but with parentheses instead of square brackets.


7 What are the advantages of using generators over regular functions?
- Generator functions have several advantages over regular functions, including:
- Memory efficiency
Generators only evaluate code when needed, so they don't need to store all values in memory at once. This makes them especially useful for large data sets.
- Lazy loading
Generators can load data on demand, which can improve performance and reduce bandwidth usage.
- Infinite sequences
Generators can create infinite sequences by pausing and resuming. This is useful for real-time data processing.
- Readability
Generators can make code easier to read and understand by breaking up the iteration process into smaller chunks.
- Speed
Generators can be faster than regular loops, especially for large data sets.
- Flexibility
Generators offer a more flexible approach to computation than regular functions.
- Easy to implement
Generators automatically implement iter(), next(), and StopIteration

10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- 1. map()
Purpose: Transforms each element of an iterable by applying a specified function.

How it works: The function is applied to every element of the iterable, and the results are returned as a map object (which is an iterator).

- 2. filter()
Purpose: Filters elements from an iterable based on a function that returns True or False.

How it works: Only elements for which the function returns True are included in the output.

- 3. reduce()
Purpose: Combines the elements of an iterable into a single value using a specified function.

How it works: The function is applied cumulatively to the elements, reducing the iterable to a single value.

````python
#Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True
is_prime(50)

````

````python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
for num in fibonacci_generator(10):
    print(num)
````

````python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
  for i in range(exponent+1):
    yield 2**i
for num in power_of_two(5):
  print(num)
````

````python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line.

    Args:
        file_path (str): The path to the file to be read.

    Yields:
        str: The next line in the file.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line)
````

````python
#Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    """
    Convert a temperature from Celsius to Fahrenheit.

    Formula: (°C * 9/5) + 32 = °F
    """
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
temperatures_celsius = [0, 20, 30, 40, 100]

# Use map to apply the conversion to each temperature in the list
temperatures_fahrenheit = list(map(celsius_to_fahrenheit, temperatures_celsius))

# Print the converted temperatures
print("Temperatures in Celsius:", temperatures_celsius)
print("Temperatures in Fahrenheit:", temperatures_fahrenheit)
````

````python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    """
    Check if a character is not a vowel.
    """
    vowels = "aeiouAEIOU"
    return char not in vowels

def remove_vowels(input_string):
    """
    Remove all vowels from the input string using filter().

    Args:
        input_string (str): The string to process.

    Returns:
        str: The string without vowels.
    """
    return ''.join(filter(is_not_vowel, input_string))

# Input string
input_str = "This is an example string."

# Remove vowels
result = remove_vowels(input_str)

# Output the result
print("Original string:", input_str)
print("String without vowels:", result)
````
